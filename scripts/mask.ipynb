```python
import numpy as np
import matplotlib.pyplot as plt
from dedalus.extras import plot_tools
```

```python
%matplotlib notebook
```

```python
Nr = 256
Nphi = 256
R = 1

r = np.linspace(0, R, Nr)[None,:]
phi = np.linspace(0, 2*np.pi, Nphi)[:,None]

x = r * np.cos(phi)
y = r * np.sin(phi)

r_mesh, phi_mesh = plot_tools.quad_mesh(r.ravel(), phi.ravel())
x_mesh = r_mesh * np.cos(phi_mesh)
y_mesh = r_mesh * np.sin(phi_mesh)
```

```python
Xinf = np.maximum(np.abs(x), np.abs(y))
f = 0.5 + 0.5*np.tanh((Xinf - R/np.sqrt(3))/0.01)

plt.figure()
plt.pcolormesh(x_mesh, y_mesh, f)
plt.axis('equal')
plt.colorbar()
```
